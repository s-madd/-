```
a = [1, 2, 3, 4]

```



````python
def unik(a, b):
  res = a + b
  for x in res:
    if res.count(x) == 2:
      res.remove(x)
      res.remove(x)
  return res

a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

print(unik(a,b))
````

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
import time


def fight(pk, pd):
  input()
  print('3')
  time.sleep(1)
  print('2')
  time.sleep(1)
  print('1')
  time.sleep(1)
  print('!!!')
  print()

  atk_knight = int(pk['weapon'] * pk['strenght'] // 1000 * 1.5)
  atk_dragon = int(pd['atk'] * 100 // pk['armor'] // 10 * 0.8)
  hp_knight = pk['health']
  hp_dragon = pd['hp']

  print('Воин:', hp_knight)
  print('Дракон:', hp_dragon)
  print()
  time.sleep(1)

  while (hp_knight > 0) and (hp_dragon > 0):

    hp_knight -= atk_dragon
    print('Воин:', hp_knight, '/ -' + str(atk_dragon))
    print('Дракон:', hp_dragon)
    print()
    time.sleep(1)

    if hp_knight > 0:

      hp_dragon -= atk_knight
      print('Воин:', hp_knight)
      print('Дракон:', hp_dragon, '/ -' + str(atk_knight))
      print()
      time.sleep(1)

    else: break

  if hp_knight > hp_dragon:
    return True
  else:
    return False


def dragon_call():

  print()
  print('Ну в общем дракон это часть воображения нашего шизанутого рыцаря. Поэтому его мы тоже сами придумаем')
  name_dragon = input('Введите имя нашего дракона: ')
  points = 5
  print()
  print('теперь у вас 5 поинтов на характеристики')

  print()
  print('Здоровье: ')
  print('1 - чешуйка, как у карася (1 point)')
  print('2 - среднестатистический дракончик (2 point)')
  print('3 - это че за.. (3 point)')

  while True:
    vvod = input('Введите цифру: ')
    if vvod == '1' or vvod == '2' or vvod == '3':
      break

  if vvod == '1':
    points -= 1
    dragon_hp = 30

  if vvod == '2':
    points -= 2
    dragon_hp = 60

  if vvod == '3':
    points -= 3
    dragon_hp = 100

  print(f'({points}p) Урон:')
  print('1 - настолько огнедышащий, что может подкурить (1 points)')
  print('2 - ну норм так (2 points)')
  print('3 - барбекю (3 points)')

  while True:
    vvod = input('Введите цифру: ')
    if vvod == '1' or vvod == '2' or vvod == '3':
      if int(vvod) <= points:
         break
      else: print('Не хватает очков')

  if vvod == '1':
    points -= 1
    dragon_atk = 30

  if vvod == '2':
    points -= 2
    dragon_atk = 60

  if vvod == '3':
    points -= 3
    dragon_atk = 100

  params = {
      'dragon': {
          'name': name_dragon,
          'hp': dragon_hp,
          'atk': dragon_atk
      }
  }

  return params['dragon']


def knight_call():

  print()
  name_knight = input('Введите имя своего воина: ')
  points = 10
  print()
  print('У вас есть 10 очков, чтобы выбрать характеристики', name_knight)

  print()
  print('Доспехи:')
  print('1 - деревянные (1 point)')
  print('2 - медные (2 point)')
  print('3 - стальные (3 point)')

  while True:
    vvod = input('Введите цифру: ')
    if vvod == '1' or vvod == '2' or vvod == '3':
      break

  if vvod == '1':
    points -= 1
    armor = 30

  if vvod == '2':
    points -= 2
    armor = 65

  if vvod == '3':
    points -= 3
    armor = 100


  print()
  print(f'({points}p) Оружие:')
  print('1 - кулаки (1 point)')
  print('2 - рогатка (2 point)')
  print('3 - меч (3 point)')

  while True:
    vvod = input('Введите цифру: ')
    if vvod == '1' or vvod == '2' or vvod == '3':
      break

  if vvod == '1':
    points -= 1
    weapon = 30

  if vvod == '2':
    points -= 2
    weapon = 65

  if vvod == '3':
    points -= 3
    weapon = 100


  print()
  print(f'({points}p) Силовые показатели (урон):')
  print('1 - слабак (1 point)')
  print('2 - атлет (2 point)')
  print('3 - Илья Муромец (3 point)')

  while True:
    vvod = input('Введите цифру: ')
    if vvod == '1' or vvod == '2' or vvod == '3':
      break

  if vvod == '1':
    points -= 1
    strenght = 30

  if vvod == '2':
    points -= 2
    strenght = 65

  if vvod == '3':
    points -= 3
    strenght = 100


  print()
  print(f'({points}p) Здоровье:')
  print('1 - дохляк (1 point)')
  print('2 - обычный крестьянин (2 point)')
  print('3 - кто ты, воин.. (3 point)')

  while True:
    vvod = input('Введите цифру: ')
    if vvod == '1' or vvod == '2' or vvod == '3':
      if int(vvod) <= points:
         break
      else: print('Не хватает очков')

  if vvod == '1':
    health = 30
    points -= 1

  if vvod == '2':
    health = 65
    points -= 2

  if vvod == '3':
    health = 100
    points -= 3

  params = {
      'knight': {
          'name': name_knight,
          'armor': armor,
          'weapon': weapon,
          'strenght': strenght,
          'health': health
      }
  }

  return params['knight']


def main():

  while True:

    print('Однажды жил был рыцарь... И дракон...')
    knight_par = knight_call()
    print(knight_par)
    print()
    print('...')
    print('(Нажимай Enter, чтобы продвигаться по сюжету)')
    print()
    print('Я всю жизнь тренировался, чтобы сегодня сделать это...')
    input()
    print('настал судный день...')
    input()
    print()
    print('*взять снаряжение*')
    input()
    print('--вы взяли снаряжение')
    print()
    print('У меня еще достаточно времени, прежде чем дракон сожжет всё королевство до тла')
    input()
    print('1 - пойти через лес (длинная дорога, малая опасность)')
    print('2 - пойти через деревню мертвецов (короткий путь, риски умереть)')

    while True:
      vvod = input('ввод: ')
      if vvod == '1' or vvod == '2':
        break

    if vvod == '1':
      print('Через лес, так через лес')
      print()
      input()
      print('*Вы не встретили ничего необычного*')
      input()

    if vvod == '2':
      print('Мы не ищем легких путей. Пойду через мертвячину')
      input()

      if knight_par['armor'] < 65:
        print('Хрысь!')
        input()
        print('*Из под змели вылез лучник-мертвец и пробил вашу слабенькую броню. Вы погибли*')
        input()
        break

      else:
        print('Ха-ха! Раскидываю этих слабаков налево и направо!')
        print('*Вы успешно прошли деревеню и вышли к полю*')
        input()

    print('Никогда не видел таких просторных полей...')
    input()
    print('*К вам подошел странник*')
    input()
    print('Чего тебе?')
    input()
    print('- помоги мне открыть банку огурцов')
    input()
    print('Что? Я иду вообще-то мир спа.. Ай, ладно, сейчас')
    input()

    if knight_par['strenght'] < 50:
      print('пук..')
      print('*кажется вы не смогли овладать с закруткой бабульки*')
      input()
      print('*вы перетрудились и потеряли сознание. надежд на спасение мира больше нет. попробуйте в следующей жизни :(*')
      input()
      break

    print('чпоньк! ну на, держи')
    input('- спасибо большое. бог с тобой!')
    print('...')
    input()
    print('7 дней... и 7 ночей воин пробыл в пути')
    input()
    print('впереди мгла... пепел... земля горяча...')
    input()
    print('А вот и ты, дракон! Хана тебе!')

    par_dragon = dragon_call()

    input()
    print('Да начнется бой!')

    flg = fight(knight_par, par_dragon)

    if flg:

      print()
      print('...')
      input()
      print('неужели это победа...')
      input()
      print('я сделал это! мир спасен!')
      input()

      break

    else:

      print()
      print('...')
      input()
      print('рыцарь погиб. миру конец...')
      break

  print()

main()
````

# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests as req


def get_pokemon(pokemons):
    name = input('Введите имя покемона (с маленькой буквы), которого хотите добавить: ')
    response = req.get(f'https://pokeapi.co/api/v2/pokemon/{name}')

    if response.status_code == 200:
        data = response.json()

        name = data['name']
        weight = data['weight']
        height = data['height']
        tipe = data['types'][0]['type']['name']
        message1 = f'{name} - Тип: {tipe}. Вес: {weight}. Рост: {height}.'
        abilities = []

        for i in data['abilities']:
            abilities.append(i['ability']['name'])

        message2 = 'Способности:', *abilities

        if name not in pokemons:
            pokemons[name] = {'info': {}}

        pokemons[name]['info']['ab1'] = message1
        pokemons[name]['info']['ab2'] = message2

        print('Покемон добавлен')
    else:
        print(f'Ошибка сервера, возможно, что такого покемона нет: {response.status_code}')


def pokemon_delete(pokemons):
    name = input('Введите имя покемона: ')

    if name in pokemons:
        del pokemons[name]

        print('Покемон исключён из команды')
    else:
        print('Такого покемона нет в команде')


def pokemon_search(pokemons):
    name = input("Ввдите имя покемона: ")

    if name in pokemons:
        print(f'''
        Имя покемона: {name}
        Информация: {pokemons[name]["info"]}
              ''')
    else:
        print('Такого покемона нет в команде')


def pokemons_info(pokemons):
    for pok in pokemons:
        print(f'''Покемон: {pok}
        Информация: {pokemons[pok]['info']['ab1']}
                    {pokemons[pok]['info']['ab2']}
        ''')


def pokemon_battle(pokemons):
    name1 = input('Введите имя первого покемона: ')
    name2 = input('Введите имя второго покемона: ')

    if name1 not in pokemons:
        print(f'Покемон {name1} не найден в команде.')
        return
    if name2 not in pokemons:
        print(f'Покемон {name2} не найден в команде.')
        return

    weight1 = int(pokemons[name1]['info']['ab1'].split('Вес: ')[1].split('.')[0])
    height1 = int(pokemons[name1]['info']['ab1'].split('Рост: ')[1].split('.')[0])

    weight2 = int(pokemons[name2]['info']['ab1'].split('Вес: ')[1].split('.')[0])
    height2 = int(pokemons[name2]['info']['ab1'].split('Рост: ')[1].split('.')[0])

    score1 = weight1 + height1
    score2 = weight2 + height2

    if score1 > score2:
        print(f'Победил {name1}!')
    elif score2 > score1:
        print(f'Победил {name2}!')
    else:
        print('Ничья!')


pokemons = {}
command = None
while command != 0:
    print('''
    1 - Добавить покемона в команду
    2 - Удалить покемона из команды
    3 - Посмотреть информацию о каждом покемоне в команде
    4 - Найти покемона по имени (вывод есть ли покемон
    в команде и информацию о нём)
    5 - Устроить тренировочный бой
    0 - Завершить программу
    ''')

    command = int(input('Введите цифру: '))

    if command == 0:
        print('''
        Завершить программу''')
    elif command == 1:
        print('''
        Добавить покемона в команду
        ''')

        get_pokemon(pokemons)
    elif command == 2:
        print('''
        Удалить покемона из команды
        ''')

        pokemon_delete(pokemons)
    elif command == 3:
        print('''
        Посмотреть информацию о каждом покемоне в команде
        ''')

        pokemons_info(pokemons)
    elif command == 4:
        print('''
        Найти покемона по имени (вывод есть ли покемон
        в команде и информацию о нём)
        ''')

        pokemon_search(pokemons)
    elif command == 5:
        print('''
        Устроить тренировочный бой
        ''')

        pokemon_battle(pokemons)
    else:
        print('''
        Введите число из списка функций
        ''')

        continue
````
